```python
usd_pkr=280.5
eur_usd=1.085
eur_pkr=307.2
```

```python
initial_capital_usd=1000000
```

```python
implied_eur_pkr=eur_usd*usd_pkr
print(implied_eur_pkr)
```

```python
# we will use 0.01 paisa as a threshold for arbitrage, however threshold should be what covers transaction fees 
if (implied_eur_pkr-eur_pkr)>0.01:
    print("Arbitrage opportunity exists")
```

```python
# Since we have USD, the trading paths can be:
#1) USD -> EUR -> PKR
#2) USD -> PKR -> EUR
```

```python
# path 1
amount=initial_capital_usd
amount = amount/eur_usd # convert to usd to eur
amount = amount*eur_pkr # convert to eur to pkr
amount = amount/usd_pkr # convert to pkr to usd
profit_1 = amount - initial_capital_usd
print("Profit from path 1: From USD - EUR - PKR - USD ", profit_1)
```

```python
#Path 2
amount=initial_capital_usd
amount = amount*usd_pkr # convert to usd to pkr
amount = amount / eur_pkr # convert to pkr to eur
amount = amount * eur_usd # convert to eur to usd
profit_2 = amount - initial_capital_usd
print("Profit from path 2: From USD - PKR - EUR - USD ", profit_2)
```

```python
if profit_1 > 0 and profit_1 > profit_2:
    print("Best path is USD - EUR - PKR - USD with profit ", profit_1)
elif profit_2 > 0 and profit_2 > profit_1:
    print("Best path is USD - PKR - EUR - USD with profit ", profit_2)
else:
    print("No profitable arbitrage opportunity exists")
```
